```python
# b) Sl책 ihop kolumnerna "책r" och "veckonummer" till en kolumn med namn "Vecka"

veckodata_riket["Vecka"] = veckodata_riket["책r"].astype(str) + "v" + veckodata_riket["veckonummer"].astype(str)

first_column = veckodata_riket.pop('Vecka')
veckodata_riket.insert(0, 'Vecka', first_column)

veckodata_riket.drop(columns=["책r", "veckonummer"], inplace=True)
veckodata_riket
```

```python
# c) Rita en linjediagram som visar avlidna fall per vecka fr책n 2020v6 till nu

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

ax = sns.lineplot(x="Vecka", y="Antal_avlidna_vecka", data=veckodata_riket)

plt.xticks(rotation=45)

# Improve the spacing between the x-axis labels, Code from ChatGPT
ax.xaxis.set_major_locator(plt.MaxNLocator(10))

plt.show()
```
