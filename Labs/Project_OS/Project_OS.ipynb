## Age distribution for participants between 18-45 years

```python
import pandas as pd
import plotly.express as px

athlete_events = pd.read_csv('../Project_OS/athlete_events.csv')

# Filter for ages between 18-45
athlete_age_range = athlete_events[(athlete_events['Age'] >= 18) & (athlete_events['Age'] <= 45)]

fig = px.histogram(athlete_age_range, 
                   x='Age', 
                   nbins=45, 
                   title='Age distribution for participants between 18-45 years',
                   template="simple_white")

fig.update_traces(hovertemplate='Age: %{x}<br>Number of participants: %{y}') # Code from ChatGPT
fig.update_layout(yaxis_title="Number of participants")

fig.show()
```
