**Use the "autos" dataset to answer the following questions:**

**Q1:** How many cars from Jaguar are in the dataset, and what is the price of the most expensive one?

<details>
<summary>Answer</summary>
<br>
&nbsp;&nbsp;&nbsp;count=3
&nbsp;&nbsp;&nbsp;price=36000.0
</details>

```python
autos.dropna(subset=["price"], inplace=True)
autos.sort_values(by="price", ascending=False, inplace=True)

number_of_jaguars = len(autos.query("make == 'jaguar'"))
most_expensive_jaguar_price = autos[autos['make'] == 'jaguar']['price'].iloc[0]

print(f"Number of Jaguar cars: {number_of_jaguars}")
print(f"Price of the most expensive Jaguar: {most_expensive_jaguar_price}")
```

**Q4:** What is the lowest price per horsepower in the dataset, and what brand ("make") is that car?

<details>
<summary>Answer</summary>
<br>
&nbsp;&nbsp;&nbsp;price per horsepower=72.84
&nbsp;&nbsp;&nbsp;brand=Toyota
</details>

```python
autos['price per hp'] = autos['price'] / autos['horsepower']
autos.dropna(subset=["price per hp"], inplace=True)
autos.sort_values(by="price per hp", ascending=True, inplace=True)

lowest_price_per_hp_make = autos['make'].iloc[0]
lowest_price_per_hp = autos['price per hp'].iloc[0]

print(f"The car with the lowest price per horsepower is a {lowest_price_per_hp_make} and it's price per hp is {lowest_price_per_hp}.")
```
