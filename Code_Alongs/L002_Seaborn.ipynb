```python
import seaborn as sns

tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.relplot(x="total_bill", y="tip", hue="sex", style="smoker", size="size", sizes=(20, 100), row="time", height=2, col="day", data=tips)
```

```python
sns.catplot(x="day", y="total_bill", kind="strip", data=tips)
```

```python
sns.displot(x="total_bill", kind="kde", data=tips)
```

```python
sns.displot(x="total_bill", bins=20, kind="hist", data=tips)
```
